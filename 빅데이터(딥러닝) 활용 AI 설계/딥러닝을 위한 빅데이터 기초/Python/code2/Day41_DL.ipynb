#### LSTM을 이용한 개체명 인식 모델 
[단어] [품사 태깅] [청크 태깅] [개체명 태깅]
#### 1. BIO : 개체명을 인식하는 방법
- B : 개체명이 시작되는 부분, I : 개체명 내부 부분, O: 개체명이 아닌 부분  
ex) BTS의 눈물을 듣자  
    BIO : BIIOOBIOOOO   

```
import re
from collections import Counter
vocab = Counter()
```

```
sentence = []
sentences = []
ner_set = set() # 개체명 태깅 집합
f = open('/content/drive/My Drive/Colab Notebooks/res/train.txt','r')

for line in f:
    if len(line)==0 or line.startswith('-DOCSTART') or line[0]=='\n':# 빈 라인이거나, txt 파일의 첫 줄이거나, line의 시작이 개행문자라면
        if len(sentence) > 0: # 
            sentences.append(sentence) # 문단 단위로 sentence 형성
            sentence=[]
        continue
    splits = line.split(" ") # line을 공백으로 split
    splits[-1] = re.sub('\n','',splits[-1]) # split한 결과의 마지막이 개행 문자라면 삭제
    
    word = splits[0].lower() # split한 결과의 처음(단어)을 소문자로 변환
    vocab[word] = vocab[word]+1 # 단어 빈도수 조사
    
    sentence.append([word, splits[-1]]) # [단어, 개체명 태깅]
    ner_set.add(splits[-1]) 
sentences
```

```
# 개체명 태깅 집합
ner_set
```

```
# 빈도수로 정렬
vocab_sorted = sorted(vocab.items(), key=lambda x:x[1], reverse=True)
```

```
vocab_sorted
```

```
# 정수 인코딩
# 2번부터 시작
word_to_index = {w: i+2 for i, (w,n) in enumerate(vocab_sorted) if n>50} # 순서, (단어, 빈도수)
len(word_to_index)
```

```
word_to_index
```

```
word_to_index["go"]
```

```
# 개체명 태깅(ner_set) 정수 인코딩
i=1
ner_to_index = {}
ner_to_index['PAD'] = 0
for ner in ner_set:
    ner_to_index[ner] = i
    i += 1
print("ner_to_index", ner_to_index) # ydata

# word_to_index의 0,1번 자리에 pad, oov로 채움
word_to_index["PAD"]=0
word_to_index["OOV"]=1
print("word_to_index", word_to_index) # xdata

# word_to_index에는 최소 빈도수가 50번을 넘는 단어들만 저장되어 있음. 약 1540개 단어가 저장되어 있음
# 0번:PAD, 1번:OOV,..., 1540:'단어'
```

```
# ner_to_index: ydata
# word_to_index: xdata (1540개 단어)
# sentences # 14000여개 문장, 21009 단어

datax = []
for s in sentences:
    tempx = []
    for w, label in s:
        try:
            tempx.append(word_to_index.get(w, 1)) # 단어 빈도수가 50이 넘는 단어인 경우 단어의 인덱스를 가져온다
        except keyError:
            tempx.append(word_to_index['OOV']) # 단어 빈도수가 50이 넘지 않는 단어인 경우
    datax.append(tempx)
print(datax)
```

```
from keras.models import Sequential
from keras.layers import *
from keras.optimizers import Adam
```

```
nwords = len(word_to_index) # 1542 단어의 종류
nlabels = len(ner_to_index) # 10 태깅 종류
```

```
model = Sequential()
model.add(Embedding(input_dim=nwords, 
                    output_dim=16, 
                    input_length=maxLen, 
                    mask_zero=True)) # padding을 해서 0이 많은 상태에서 0을 연산에서 빼겠다 →입력 용도로만 0을 사용하고 연산X
model.add(Bidirectional(LSTM(32, return_sequences=True))) # 양방향 LSTM 사용
model.add(Dense(nlabels, activation="softmax"))
# return_sequences = True: many to many 매 셀마다 출력되는 구조
# return_sequences = False: many to one
```

```
model.compile(loss="categorical_crossentropy", optimizer=Adam(0.01), metrics=["accuracy"])
```

```
from keras.utils import np_utils
ytrain2 = np_utils.to_categorical(ytrain)
ytrain2[0][0]
```
