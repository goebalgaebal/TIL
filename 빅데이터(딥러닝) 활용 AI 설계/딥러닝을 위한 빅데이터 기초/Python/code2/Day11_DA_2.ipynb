### 이미지 저장

```python
# 이미지 저장 1
# url을 통해 읽어온 후 바로 하드디스크에 저장
import urllib.request
url = "https://www.google.co.kr/images/branding/googlelogo/1x/googlelogo_color_272x92dp.png"
sname = "test.png" # 저장 파일 이름
urllib.request.urlretrieve(url, sname) # urllib.request.urlretrieve(접속 주소, 저장 파일 이름)
print("저장 완료")
```

```python
# 이미지 저장 2
# url을 통해 읽어온 후 메모리에 저장, 그 다음 메모리에 있는 데이터를 파일로 저장
import urllib.request
url = "https://www.google.co.kr/images/branding/googlelogo/1x/googlelogo_color_272x92dp.png"
sname = "test2.png" # 저장 파일 이름
mem = urllib.request.urlopen(url).read() # 접속(통신을 위한 객체가 생성됨) 후 읽어오기 
mem
```

```python
with open(sname, mode="wb") as f:
    f.write(mem)
    print("저장 완료")
```

## find_all()을 이용한 여러 요소

```python
html="""
<html><body>
<ul>
<li><a href="http://www.naver.com">네이버</a></li>
<li><a href="http://www.daum.net">다음</a></li>
</ul>
</body></html>
"""
```

```python
soup = BeautifulSoup(html, "html.parser")
soup.find_all("ul")
```

```python
soup.find_all("li")
```

```python
links = soup.find_all("a")
links
```

```python
for link in links:
    href = link.attrs["href"] # a = tag, href = attribute
    text = link.string # tag로 묶인 값
    print(text, "→", href)
```

기상청 데이터에서
1. title tag 내용 추출
2. wf tag 내용 추출
3. tmn, tmx 값 추출 후 DF로 저장

```python
import urllib.request
import urllib.parse
from bs4 import BeautifulSoup

values = {"stnId":"184"} # 제주도
parames = urllib.parse.urlencode(values) 

url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
data = urllib.request.urlopen(url).read()
html = data.decode("utf-8")
print(html)
```

```python
soup = BeautifulSoup(html, "html.parser")
title = list(map(lambda x: x.string, soup.find_all("title")))
title
```

```python
wfs = soup.find_all("wf")
wfStr = []
for wf in wfs:
    wfStr.append(wf.string)
wfStr
```

```python
tmn_tmx = {"tmn":[], "tmx": []}

taglist = ["tmn", "tmx"]
for tag in taglist:
    res = soup.find_all(tag)
    
    for content in res:
        tmn_tmx[tag].append(content.string)
print(tmn_tmx)
print("\ntmn 길이", len(tmn_tmx["tmn"]), "\ntmx 길이", len(tmn_tmx["tmx"]))
```

```python
import pandas as pd
df = pd.DataFrame(tmn_tmx)
df.info()
df.head()
```

## CSS 선택자를 사용한 추출
jQuery처럼 css선택자 사용가능, 특정 요소 추출 <br><br>
* soup.select(<선택자>): 여러개 요소
* soup.select_one(<선택자>): 1개 요소 <br><br>

* id:#id로 구분
* class: .class로 구분

```python
html = """
<html><body>
<div id="myid">
<h1>멀티캠퍼스</h1>
<ul class="items">
<li>강의실</li>
<li>엘리베이터</li>
<li>식당</li>
</ul>
</div></body></html>
"""
```

```python
soup = BeautifulSoup(html, "html.parser")
soup.select_one("h1")
```

```python
soup.select_one("li")
```

```python
soup.select("li")
```
