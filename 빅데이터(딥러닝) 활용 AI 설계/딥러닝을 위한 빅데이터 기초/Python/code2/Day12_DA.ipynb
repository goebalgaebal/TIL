## CSS 선택자 추출

```python
html = """
<ul id="books">
<li id="b1">상실의 시대</li>
<li id="b2">데미안</li>
<li id="b3">이방인</li>
<li id="b4">태백산맥</li>
<li id="b5">어린왕자</li>
</ul>
"""
```

```python
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
# 람다함수 이름: sel, 람다함수 인자: q
# sle 함수를 호출할 때 전달되는 인자(css 선택자)로부터 soup에 저장된 해당 텍스트를 추출 
sel = lambda q: print(soup.select_one(q).string)
```

```python
sel("#b4")
sel("li#b4")
sel("ul li#b4")
sel("ul > li#b4")
sel("#books #b4")
sel("#books > #b4")
sel("li[id='b4']")
sel("li") # 여러개가 있는 경우 select_one이기 때문에 처음 1개만 출력됨
sel("li:nth-child(2)")
sel("li:nth-of-type(2)")
```

#question-summary-57067264 > div.summary > h3 > a
#question-summary-57067260 > div.summary > h3 > a

```python
page = urlopen("https://stackoverflow.com/questions/tagged/python")
document = page.read()
document
```

```python
soup = BeautifulSoup(document, "html.parser")
questions = soup.find(id="questions")
questions
```

```python
questions.find_all("a")
```

```python
questions_list = questions.find_all("a", class_="question-hyperlink")
questions_list
```

```python
for questions in questions_list:
    print(questions.get_text())
    # print(questions.string)
    print("http://stackoverflow.com/" + questions.get("href"))
```

### Top box office

```python
url = "https://www.rottentomatoes.com/"
soup = BeautifulSoup(req.urlopen(url).read(), "html.parser")
soup
```

```python
top_box_office = soup.select("#homepage-top-box-office td.middle_col a")
top_box_office
```

```python
for movie in top_box_office:
    print(movie.string)
```
