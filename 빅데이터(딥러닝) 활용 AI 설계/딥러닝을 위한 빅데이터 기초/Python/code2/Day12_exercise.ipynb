### stackoverflow 사이트에서 'python'키워드로 검색했을 때 출력되는 질문 추출
(답변x=>'답변이 아직 없네요' 출력, 답변o(모든 답변 출력))

```python
url = "https://stackoverflow.com/questions/tagged/python/"
res = req.urlopen(url)
soup = BeautifulSoup(res, "html.parser")
soup
```

```python
questions = soup.find_all("div", class_="question-summary")
print("질문 개수:", len(questions))
print("질문 형식")
questions[0]
```

```python
questions[0].select(".status.unanswered")
```

```python
head_adrr = "https://stackoverflow.com/"
for idx, q in enumerate(questions):
#     print(q.select(".status.unanswered"))
    print("="*149)
    res = q.select(".status.answered")
    if res != []: # 질문에 답변이 있으면
        anum = res[0].select("strong")[0].string # 답변 개수
        print("[", idx+1, "번 질문 답변", anum, "개]")
    else:
        print("[", idx+1, "번 질문 답변 X]")

    tail_adrr = q.select(".question-hyperlink")[0]
    print(tail_adrr.string)
    qlink = head_adrr + tail_adrr.get("href")
    print("질문 링크", qlink)

    qres = req.urlopen(qlink) # 질문과 답변이 있는 페이지의 html 읽기
    qsoup = BeautifulSoup(qres, "html.parser")
    print("질문 내용\n", qsoup.select("#question .post-text")[0].get_text()) # 질문 내용
    
    if res == []: # 질문에 답변이 없으면
        continue
    else:
        answers = qsoup.select("div#answers .answer .post-text") # 답변을 담고있는 list
        for idx, answer in enumerate(answers):
            print(idx+1, "답변\n", answer.get_text()) 
```

```python
from requests import get
import re

for page in range(1, 101, 10):
    print("#"*149)
    print("Page", (page//10)+1)
    purl = "https://search.naver.com/search.naver?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%EB%A8%B8%EC%8B%A0%EB%9F%AC%EB%8B%9D&c_id=&c_name=&sm=tab_pge&kin_start={}".format(page)

    # 페이지의 html 정보 받아오기
    res = req.urlopen(purl)
    soup = BeautifulSoup(res, "html.parser")
    titles = soup.select(".question a") # 페이지의 각 질문글 제목만 추출
    questions = soup.select(".question") # 페이지의 각 질문글 추출
    
    for idx, q in enumerate(questions):
        qurl = q.select("a")[0].get("href") # 각 질문의 link 추출
        # print(qurl)
        qres = get(qurl)
        qsoup = BeautifulSoup(qres.text, "html.parser")
    
        # 질문글 제목 출력
        # 질문글의 제목이 없는 경우도 있기 때문에 페이지에서 제목을 가져왔습니다
        print("="*149)
        print("[ 질문 제목 : ",titles[idx].text, " ]")
        qcont = qsoup.select(".question-content .c-heading__content")
    
        # 질문의 내용을 가져올 때 총 3가지 경우의 수가 있었습니다
        # 아마 검색의 키워드에 따라서 다를거라고 생각합니다
        # 1. 질문의 내용이 따로 있는 경우
        # 2. 질문의 제목과 내용이 같이 있는 경우
        # 3. 나이에 따라 질문글을 볼 수 없는 경우
        if qcont != []:
            print(qsoup.select(".question-content .c-heading__content")[0].get_text().strip())
        else:
            if qsoup.select(".title") != []:
                print(qsoup.select(".question-content .title")[0].get_text().strip())
            else:
                ecase = qsoup.select("script")[0].get_text()
                e = re.search("window.alert\((.*)\);", ecase)
                print("Error : ", e.group(1))
        
        print("-"*149)
        acont = qsoup.select(".answer-content__list._answerList ._endContentsText.c-heading-answer__content-user")  
        if acont != []:
            for idx, answer in enumerate(acont):
                print("[ ", idx+1, "번째 답변 ]")
                print(answer.get_text().strip(), "\n")
    print("\n")
  
```

```python
def soup_url( url ):
    data = req.urlopen(url).read()
    data.decode(encoding = "utf-8", errors = "ignore")
    return BeautifulSoup(data, "html.parser")

# 크롤링할 페이지 갯수
pageNo = 10
 
# 네이버 클레오파트라 검색 페이지에 대한 soup객체 리스트
soups = [ soup_url("https://kin.naver.com/search/list.nhn?query=%ED%81%B4%EB%A0%88%EC%98%A4%ED%8C%8C%ED%8A%B8%EB%9D%BC&page=" + str(i))   for i in range(1, pageNo + 1 ) ]

threadUrls = []
for soup in soups:
    threadUrls += [ tag.attrs['href'].strip().split("§")[0] for tag in soup.select("#s_content > div.section > ul > li > dl > dt > a")]
 
soups_InPage = [ soup_url(thrad_url) for thrad_url in threadUrls ]
 
titleSelector     = r"div.title"
questionSelector  = r"div.c-heading__content"
answerSelector    = r"div.c-heading-answer__content-user"
 
titles    = []
questions = []
answers   = []
 
# 연령 인증이 걸려있는 질문글은 건너뛰었습니다
for soup in soups_InPage:
    if soup.select_one(titleSelector) != None:
        titles.append(soup.select_one(titleSelector).text.strip() )
        questions.append(soup.select_one(questionSelector).text.strip() if soup.select_one(questionSelector) != None else "" )
        answers.append(soup.select_one(answerSelector).text.strip() if soup.select_one(answerSelector) != None else "")

print(len(titles) , len(questions), len(answers))
 
print(titles[:10]) 
print(questions[:10]) 
print(answers[:10])
```

### 1부터 10,000까지 8이라는 숫자가 총 몇번 나오는가? <br>
8이 포함되어 있는 숫자의 갯수를 카운팅 하는 것이 아니라 8이라는 숫자를 모두 카운팅 해야 한다. <br>
(※ 예를들어 8808은 3, 8888은 4로 카운팅 해야 함 <br><br>

1부터 10000까지 8의 개수 산정하는 것을 문제는 0000부터 9999까지 8의 개수 산정하는 것으로 변형 할 수 있있다 <br>


1. 8이 하나 일때, 4자리 중에 8이 하나고 8을 제외한 0~9까지 숫자가 3번 반복 <br>
8000 0800 0080 0008
$$_4C_1 * 9 ^ 3 * 1= 2916$$

1. 8이 두개 일때, 4자리중에 8이 두개고 두개의 8을 제외한 숫자가 2번 반복 <br>
8800 8080 8008 0880 0808 0088
$$_4C_2 * 9 ^ 2 * 2 = 972$$

1. 8이 세개 일때, 4자리중에 8이 세개고 세개의 8을 제외한 숫자가 한번 반복되므로 <br>
8880 8088 8808 0888
$$_4C_3 * 9 ^ 1 * 3 = 108$$

1. 8이 네개 일때, 4자리중에 8이 네개고 네개의 8을 제외한 숫자가 필요 없으므로 <br>
8888
$$_4C_4 * 9 ^ 0 * 4 = 4$$
<br><br>
2916 + 972 + 180 + 4 = 4000 <br>
$$ n자리수 에서 8이 등장하는 횟수 = \sum_{i=1}^{n} nC_i * 9 ^{ ( n - i )} * i $$

---
중복된 숫자를 셈으로써 8이 포함되어 있는 숫자의 개수를 세는 것이 아니라 8이라는 숫자의 등장 횟수를 센다 
1. 0부터 9까지에서 8의 빈도 계산 → 1
1. 00부터 99까지 8의 빈도 계산 → 20
 - 10의 자리에 들어가는 8의 빈도 계산: 80-89 → 10
 - 1의 자리에 들어가는 8의 빈도 계산
   각 10의 자리수에 해당하는 수마다 1의 자리에 들어오는 8의 개수는 하나밖에 없기 때문에 08, 18, 28, ... 98 → 10

1. 000부터 999까지 8의 빈도 계산 → 300.
 - 100의 자리수에 들어가는 8의 빈도 계산: 800-899 → 100
 - 10의 자리수에 들어가는 8의 빈도 계산 : 백의 자리(0-9) 8 일의 자리(0-9)  → 100
 - 1의 자리수에 들어가는 8의 빈도 계산 : 백의 자리(0-9) 십의 자리(0-9) 8 → 100

1. 0000부터 9999까지 들어가는 8의 빈도 계산 → 4000
 - 3번의 계산과 유사하게 각 자리수는 천번씩 들어오고 4자리가 있으므로 1000*4

 $$n자리수(0 - 999...999가 n번 반복) 에서 8이 등장하는 횟수 =n*10^{(n-1)} $$
