### MultiIndex.from_tuples() 
tuple로부터 multi index를 만들어주는 함수

```python
mi = pd.MultiIndex.from_tuples([("c1", "2019"), 
                                ("c1", "2020"), 
                                ("c2", "2021"), 
                                ("c2", "2022")])
mi
```

## Pandas melt()

```python
df = pd.DataFrame({"cid": ["c1", "c1", "c2", "c2"],
                   "pcd": ["p1", "p2", "p1", "p2"],
                   "pcnt": [1, 2, 3, 4],
                   "pamt": [10, 20, 30, 40]})
df
```

```python
pd.melt(df)
```

```python
pd.melt(df, id_vars=["cid", "pcd"])
```

```python
pd.melt(df, id_vars=["cid", "pcd"], var_name="pname", value_name="vname")
```

## Pandas crosstab()

```python
df = pd.DataFrame({"id": ["id1", "id1", "id1", "id2", "id2", "id3"],
               "d1": ["a", "a", "a", "b", "b", "b"],
               "d2": ["d", "d", "d", "c", "c", "c"]})
df
```

```python
pd.crosstab(index=df.d1, columns=df.d2)
```

```python
pd.crosstab(df.id, df.d1)
```

## 시각화 패키지 seaborn

```python
import seaborn as sns
```

```python
anscombe = sns.load_dataset("anscombe") # sample dataset 평균, 분산, 상관관계가 같은 dataset들(통계의 함정)
anscombe
```

```python
dataset_1 = anscombe[anscombe["dataset"] == "I"]
dataset_1
```

```python
plt.plot(dataset_1["x"], dataset_1["y"])
```

```python
plt.plot(dataset_1["x"], dataset_1["y"], "o")
```

```python
dataset_1 = anscombe[anscombe["dataset"] == "I"]
dataset_2 = anscombe[anscombe["dataset"] == "II"]
dataset_3 = anscombe[anscombe["dataset"] == "III"]
dataset_4 = anscombe[anscombe["dataset"] == "IV"]
```

```python
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white')

axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)
```

figure(): 그래프 출력 객체 <br>
figure().add_subplot(row 크기, col 크기, 순서) : 서브 영역을 나눠서 구역을 할당하는 함수

```python
axes1.plot(dataset_1["x"], dataset_1["y"], "o")
axes2.plot(dataset_2["x"], dataset_2["y"], "o")
axes3.plot(dataset_3["x"], dataset_3["y"], "o")
axes4.plot(dataset_4["x"], dataset_4["y"], "o")
fig
```

```python
axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")
fig.suptitle("Anscombe Data")
fig.tight_layout()
fig
```
