## 토지1, 전자파일 에서 대화 추출하기
[[출처] 국립국어원 언어정보나눔처](https://ithub.korean.go.kr/user/total/database/corpusView.do?boardSeq=2&articleSeq=2080&boardGb=T&isInsUpd=&boardType=CORPUS)

```python
before, after = [], []
with open("./res/BEXX0003.txt", "rt", encoding="UTF16") as inf:
    text = BeautifulSoup(inf.read(), "html.parser")
    plist = text.find_all("p") # <p> tag 추출
    
    pattern = re.compile("\"[\w\D]+\"|\'[\w\D]+\'") # 대화, 혼자말 패턴
    for line in plist:
        before.append(line.string)
        
        res = pattern.search(line.string)
        if res != None:
            after.append(res.group(0))
        else:
            after.append(None)
beforeDf = pd.DataFrame(before, columns=["before"])
afterDf = pd.DataFrame(after, columns=["after"])
```

```python
before, after = [], []
with open("./res/BEXX0003.txt", "rt", encoding="UTF16") as inf:
    raw = inf.read()
    
    plist = re.findall("<p>.+<p>", raw) # <p> tag 추출
#     plist = re.findall("<p>[\w\D]*?<\/p>", raw) # <p> tag 추출
    
    pattern = re.compile("\"[\w\D]+\"|\'[\w\D]+\'") # 대화, 혼자말 패턴
    for line in plist:
        before.append(line[3:-4])
        
        res = pattern.search(line[3:-4])
        if res != None:
            after.append(res.group(0))
        else:
            after.append(None)
beforeDf = pd.DataFrame(before, columns=["before"])
afterDf = pd.DataFrame(after, columns=["after"])
```

```python
compareDf = pd.concat([beforDf, afterDf], axis=1)
compareDf.head(30)
```
