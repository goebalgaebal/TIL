### Dataframe 병합 Merge
* how(left, right, inner(default), outer)
* on(병합 기준 변수)

```python
dfleft = pd.DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})

dfright = pd.DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']})
```

```python
dfleft
```

```python
dfright
```

```python
pd.merge(dfleft, dfright) # 공통된 Key값으로 merge
```

```python
pd.merge(dfleft, dfright, on = "KEY")
```

```python
pd.merge(dfleft, dfright, on = "KEY", how="left") 
# left의 KEY를 기준으로 merge
# 기준이 되는 data는 전체가 출력이 된다 
```

```python
pd.merge(dfleft, dfright, on = "KEY", how="inner") 
```

```python
pd.merge(dfleft, dfright, on = "KEY", how="outer") 
```

```python
dfleft = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']},
                  index = ['K0', 'K1', 'K2', 'K3'])

dfright = pd.DataFrame({'C': ['C2', 'C3', 'C4', 'C5'],
                     'D': ['D2', 'D3', 'D4', 'D5']},
                   index = ['K2', 'K3', 'K4', 'K5'])
```

```python
dfleft
```

```python
dfright
```

```python
pd.merge(dfleft, dfright) # 공통된 KEY 없음
```

```python
pd.merge(dfleft, dfright, left_index=True, right_index=True)
# left, right에 모두 다 존재하는 index에 대한 데이터 병합을 하겠다
```

```python
pd.merge(dfleft, dfright, left_index=True, right_index=True, how="left")
```

```python
dfleft.join(dfright, how="left")
```

```python
dfleft.join(dfright, how="inner")
```

### 결측치 

```python
dfleft = pd.DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})

dfright = pd.DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']})
```

```python
dfa = pd.merge(dfleft, dfright, on="KEY", how="outer")
dfa
```

```python
pd.isnull(dfa) # dfa.isnull()
```

```python
pd.notnull(dfa)
```

```python
dfleft = pd.DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['1.1', '2.2', '3.3', '4.4']})

dfright = pd.DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']})
```

```python
dfa = pd.merge(dfleft, dfright, on="KEY", how="outer")
dfa
```

### NumPy의 NaN상수를 이용해 결측치 채우기

```python
df = pd.DataFrame(np.random.randn(5, 3),
            columns=["c1", "c2", "c3"])
df
```

```python
df.ix[0, 0] # label 명이 있지만 기본적으로 col은 0, 1, 2, ...로 접근가능
```

```python
df.ix[0, 0] = None
df
```

```python
df.ix[1, [0, 2]] = np.nan # NumPy 상수 사용
df
```

```python
df.ix[1, ["c1", "c3"]] = np.nan
df.ix[2, "c2"] = np.nan
df.ix[3, "c2"] = np.nan
df.ix[4, "c3"] = np.nan
df
```

```python
df.fillna(100)
```

```python
df.fillna(0)
```

```python
df.fillna("missing")
```

```python
df.fillna(method="ffill") # forward방향으로 진행하면서 NaN을 만나기 이전의 data로 채우기
```

```python
df.fillna(method="ffill", limit=1) # 채우는 개수 제한
```

```python
df.fillna(df.mean()) # mean으로 NaN 채우기
```

```python
df.fillna(df.mean()["c1"]) # c1의 평균으로 NaN 채우기
```

```python
df2 = pd.DataFrame({"a1": [1, 2, 3, 4, 5],
                   "a2": [10, 11, 12, 13, 14]})
df2.ix[[1, 3], ["a2"]] = np.nan
df2
```

```python
df2["a3"] = np.where(pd.notnull(df2["a2"]) == True, df2["a1"], df2["a2"])
df2
```
