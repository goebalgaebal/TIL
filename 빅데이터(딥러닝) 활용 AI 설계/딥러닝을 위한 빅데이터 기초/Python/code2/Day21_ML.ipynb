```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
x1data = [70, 90, 88, 95, 75] # 5명 3월 모의고사 [학생1, 학생2, ..., 학생5]
x2data = [80, 88, 90, 98, 65] # 5명 4월
x3data = [75, 93, 90, 100, 70] # 5명 5월
ydata = [150, 185, 180, 200, 140] # 수능 점수
# 예측 [148.11533 185.45662 180.08517 198.66158 143.19026]
```

```python
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
w1 = tf.Variable(tf.random_normal([1]), name="weight1")
w2 = tf.Variable(tf.random_normal([1]), name="weight2")
w3 = tf.Variable(tf.random_normal([1]), name="weight3")
b = tf.Variable(tf.random_normal([1]), name="bias")
```

```python
hf = x1*w1+x2*w2+x3*w3+b
cost = tf.reduce_mean(tf.square(hf-y))
```

```python
train = tf.train.GradientDescentOptimizer(1e-5).minimize(cost)
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cv, hv, _ = sess.run([cost, hf, train], feed_dict={x1: x1data, x2: x2data, x3: x3data, y: ydata})
    if step%10 == 0:
        print(step, "cost:", cv, "prediction:", hv)
```

### Queue 
FIFO(First In First Out)

## Queue Runner
데이터를 읽을 때 사용하는 저장소. 멀티 스레드 기반으로 동작. 메모리에 데이터가 올라와야 학습 데이터로 사용할 수 있다. 데이터가 너무 큰 경우는 메모리에 한꺼번에 올리기 부담이 될 수도있다. 그래서 나누어서 데이터를 읽을 때 사용
* coordinator <br>
스레드 관리
* Enqueue <br>
데이터를 Queue에 집어넣는 행위  <br>
enqueue operation: 데이터를 큐에 어떻게 집어넣을 지 정의

```python
QUEUE_LENGTH = 20 # Queue의 크기 저장될 데이터 최대 개수
q = tf.FIFOQueue(QUEUE_LENGTH, "float") # Queue생선(Queue길이, Queue에 저장될 데이터 타입)
q
```

```python
# 데이터 4개를 한번에 Queue에 집어넣어라 라고 정의
enq_ops = q.enqueue_many(([1.0, 2.0, 3.0, 4.0], )) # [1.0, 2.0, 3.0, 4.0]: Queue에 저장될 데이터
# qr이라는 node: Queue에 데이터를 집어넣기
qr = tf.train.QueueRunner(q, [enq_ops, enq_ops, enq_ops]) # 3개의 thread 어떤 순서로 동작될 지는 모른다
```

```python
sess = tf.Session()
coord = tf.train.Coordinator()
# thread 생성 
#start=True라고 하지않으면 Queue에 데이터가 저장되지 않는다
threads = qr.create_threads(sess, coord=coord, start=True) 
```

```python
for step in range(10):
    print(sess.run(q.dequeue())) # Queue에서 데이터 꺼내기
    
coord.request_stop() # thread 정지
coord.join(threads) # 정지될 때까지 기다리는 함수 = 한 thread가 끝날 때까지 다른 thread는 대기
```

```python
QUEUE_LENGTH = 20 
q = tf.FIFOQueue(QUEUE_LENGTH, "float") 
enq_ops1 = q.enqueue_many(([1.0, 2.0, 3.0, 4.0], )) 
enq_ops2 = q.enqueue_many(([5.0, 6.0, 7.0, 8.0], )) 
enq_ops3 = q.enqueue_many(([9.0, 10.0, 11.0, 12.0], )) 
qr = tf.train.QueueRunner(q, [enq_ops1, enq_ops2, enq_ops3]) 
```

```python
sess = tf.Session()
coord = tf.train.Coordinator()
threads = qr.create_threads(sess, coord=coord, start=True) 
for step in range(30):
    print(sess.run(q.dequeue()))
    
coord.request_stop()
coord.join(threads)
```

## [House Prices: Advanced Regression Techniques](https://www.kaggle.com/c/house-prices-advanced-regression-techniques)
1. sqft_living에 따른 price - 데이터(8:2) 분리
2. sqft_living, bedrooms, bathrooms에 따른 price
3. sqft_living, bedrooms, bathrooms, sqft_lot, floors, zipcode에 따른 price

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# 데이터 읽기
housing = pd.read_csv("./res/연습문제데이터(집값예측)/train.csv", index_col=0)
housing.info()
```

### 1. sqft_living에 따른 price 

```python
xdata = housing.sqft_living
ydata = housing.price
xdata = np.array(xdata).reshape((-1, 1))
ydata = np.array(ydata).reshape((-1, 1))
print(xdata.shape, ydata.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(xdata, ydata, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
x = tf.placeholder(tf.float32, shape=[None, 1]) #   x * w  = y → [None, 3][3, 2] = [None, 2]
y = tf.placeholder(tf.float32, shape=[None, 1])
w = tf.Variable(tf.random_normal([1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")
```

```python
# hf = tf.matmul(x, w) + b 
hf = x*w + b
cost = tf.reduce_mean(tf.square(hf-y))
train = tf.train.GradientDescentOptimizer(0.1).minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2501):
    cv, hv, _ = sess.run([cost, hf, train], feed_dict={x: X_train_scale, y: y_train})
    if step%100 == 0:
        print(step, "cost:", cv)
```

```python
error = []
for i, test in enumerate(X_test_scale):
    e = y_test[i] - sess.run(hf, feed_dict={x: test.reshape(-1, 1)})
    print("실제값: %.2f, 예측값: %.2f, 오차: %.2f" % (y_test[i], 
                                        sess.run(hf, feed_dict={x: test.reshape(-1, 1)}), 
                                        abs(e)))
    error.append(e)
print("평균 오차 : %.2f" % (sum(error)/len(error)))
```
