히트맵 [seaborn.heatmap](http://seaborn.pydata.org/generated/seaborn.heatmap.html?highlight=heatmap#seaborn.heatmap) <br>
* 변수사이의 관계 - 대각요소 중 한쪽만 출력, 일반적으로 아래쪽을 출력

```python
titanic_size = titanic.pivot_table(index="class", columns="sex", aggfunc="size")
titanic_size
```

[seaborn.catplot](http://seaborn.pydata.org/generated/seaborn.catplot.html?highlight=catplot#seaborn.catplot) <br>
특정 데이터의 건수를 레벨 단위로 출력


```python
titanic.survived.notnull().head()
```

```python
# 나이와 성별에 따라 생존율에 영향이 있는가? 기준 : class
data = titanic[titanic.survived.notnull()]
sns.catplot(data=data, x="age", y="sex", hue="survived", row="class", kind="violin")
```

```python
sns.catplot(data=data, x="age", y="sex", hue="survived", 
            row="class", kind="violin",
            height=3, split=True)
```

```python
sns.catplot(data=data, x="age", y="sex", hue="survived", 
            row="class", kind="violin",
            height=3, split=True,
            aspect=2)
```

```python
np.random.seed(111)
# 100X3의 난수 DF 생성
pd.DataFrame(np.random.randn(100, 3),
            index=pd.date_range("7/9/2019", periods=100),
            columns=["X1", "X2", "X3"]).tail() 
```

```python
np.random.seed(111)
# DF의 각각의 열에 cumsum()(=누적합)적용
# 시계열 데이터를 나타낼 때 유용
df = pd.DataFrame(np.random.randn(100, 3),
            index=pd.date_range("7/9/2019", periods=100),
            columns=["X1", "X2", "X3"]).cumsum()
df.tail()
```

```python
df.plot()
```
